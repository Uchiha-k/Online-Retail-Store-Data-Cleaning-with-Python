```python
# Drop rows where all cells are blank
online_store = online_store.dropna(how="all")

# Drop rows where Description is blank
online_store = online_store.dropna(subset=["Description"])

online_store
```

```python
# Drop rows where all cells are blank
online_store = online_store.dropna(how="all")

# Drop rows where Description is blank
online_store = online_store.dropna(subset=["CustomerID"])

online_store
```

```python
online_store.isnull().sum()
```

```python
online_store["Decision"] = online_store["Quantity"].apply(lambda x: "remove" if x <= 1 else "Keep")
```

```python
online_store
```

```python
online_store.columns = online_store.columns.str.strip().str.lower().str.replace(" ", "_")
online_store
```

```python
online_store["revenue"] = online_store["quantity"] * online_store["unitprice"]
online_store
```

```python
# Keep only rows where Decision == "Keep"
online_store = online_store[online_store["decision"] == "Keep"]
online_store
```

```python

# Convert InvoiceDate column to datetime
online_store["invoicedate"] = pd.to_datetime(online_store["invoicedate"])

# Keep only the date part (drop the time)
online_store["invoicedate"] = online_store["invoicedate"].dt.date

online_store
```

```python
# Save cleaned dataset to Excel
online_store.to_excel("cleaned_online_retail.xlsx", index=False)
```
